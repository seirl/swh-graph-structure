```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np

DATASET = Path('inout')
DISTRIBUTIONS = [
    'cnt_in_dir',
    'dir_in_all',
    'dir_in_dir',
    'dir_in_rev',
    'dir_out_all',
    'dir_out_cnt',
    'dir_out_dir',
    'dir_out_rev',
    'ori_out_snp',
    'rel_in_snp',
    'rev_in_all',
    'rev_in_dir',
    'rev_in_rel',
    'rev_in_rev',
    'rev_in_snp',
    'rev_out_rev',
    'snp_in_ori',
    'snp_out_all',
    'snp_out_rel',
    'snp_out_rev',
]

types_verbose = {
    'cnt': 'contents',
    'dir': 'directories',
    'rev': 'revisions',
    'rel': 'releases',
    'snp': 'snapshots',
    'ori': 'origins',
    'all': 'everything',
}

def load_distribution(name):
    d = [tuple(map(int, r.split())) for r in (DATASET / (name + '.txt')).read_text().split('\n') if r]
    x = [r[0] for r in d]
    y = [r[1] for r in d]
    return x, y


def show_distribution(name):
    src, direction, dst = name.split('_')
    title = "{} of {} {} {}".format(
        ("Indegrees" if direction == 'in' else "Outdegrees"),
        types_verbose[src],
        ("from" if direction == 'in' else "to"),
        types_verbose[dst],
    )
    
    x, y = load_distribution(name)
    y_cum = np.cumsum(y[::-1])[::-1]
    fig = plt.figure()
    
    fig.suptitle(title)
    ax = fig.add_subplot()
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('Degree')
    ax.set_ylabel('Number of nodes')
    ax.plot(x, y, '.', markersize=2, label="Frequency (= x)")
    ax.plot(x, y_cum, '.', markersize=2, label="Cumulative frequency (â‰¥ x)")
    ax.legend(markerscale=4)
```

# Contents

## Indegrees

```python
show_distribution('cnt_in_dir')
```

# Directories

## Indegrees

```python
show_distribution('dir_in_all')
show_distribution('dir_in_dir')
show_distribution('dir_in_rev')
```

## Outdegrees

```python
show_distribution('dir_out_all')
show_distribution('dir_out_cnt')
show_distribution('dir_out_dir')
show_distribution('dir_out_rev')
```

## Indegrees

```python
show_distribution('rev_in_all')
show_distribution('rev_in_dir')
show_distribution('rev_in_rel')
show_distribution('rev_in_rev')
show_distribution('rev_in_snp')
```

```python
show_distribution('rev_out_rev')
```
