# Replication Package : In/Out Degree Distributions 

```python
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
import common

DATASET = Path('../experiments/inout')
DISTRIBUTIONS = [
    'cnt_in_dir',
    'dir_in_all',
    'dir_in_dir',
    'dir_in_rev',
    'dir_out_all',
    'dir_out_cnt',
    'dir_out_dir',
    'dir_out_rev',
    'ori_out_snp',
    'rel_in_snp',
    'rev_in_all',
    'rev_in_dir',
    'rev_in_rel',
    'rev_in_rev',
    'rev_in_snp',
    'rev_out_rev',
    'snp_in_ori',
    'snp_out_all',
    'snp_out_rel',
    'snp_out_rev',
]

def show_distribution(name):
    src, direction, dst = name.split('_')
    title = "{} of {} {} {}".format(
        ("Indegrees" if direction == 'in' else "Outdegrees"),
        common.types_verbose[src],
        ("from" if direction == 'in' else "to"),
        common.types_verbose[dst],
    )
    
    x, y = common.load_text_distribution(DATASET / 'per_type' / (name + '.txt'))
    dist = common.Distribution(x, y, "Degree", "Number of nodes", title)

    common.plot_all(dist, 'figures/inout', name)
```

# Layers

```python
def show_distribution_layer(name, direction):
    title = "{} ({})".format(
        ("Indegrees" if direction == 'in' else "Outdegrees"),
        common.LAYERS[name],
    )
    
    x, y = common.load_text_distribution(DATASET / ('{}_{}.txt'.format(name, direction)))
    dist = common.Distribution(x, y, "Degree", "Number of nodes", title)
    print("Average degree of {}: {}".format(name, sum(degree * count for degree, count in zip(x, y)) / sum(y)))

    common.plot_all(dist, 'figures/inout', '{}_{}'.format(name, direction))
```

```python
show_distribution_layer('full', 'in')
show_distribution_layer('full', 'out')

show_distribution_layer('dir+cnt', 'in')
show_distribution_layer('dir+cnt', 'out')

show_distribution_layer('rev', 'in')
show_distribution_layer('rev', 'out')

show_distribution_layer('rel+rev', 'in')
show_distribution_layer('rel+rev', 'out')

show_distribution_layer('ori+snp', 'in')
show_distribution_layer('ori+snp', 'out')
```
